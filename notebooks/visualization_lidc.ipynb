```python
import torch
import os
import numpy as np
import matplotlib.animation as animation
import matplotlib.pyplot as plt
```

```python
data_folder = 'data/LIDC/LIDC-IDRI-0032/'

files = os.listdir(data_folder)
images = []
files.sort()
for file in files:
    if file.endswith('.npy'):
        data = np.load(data_folder + file)
        images.append(data)

fig = plt.figure()
plt.axis("off")
ims = [[plt.imshow(i, cmap='gray')] for i in images]
ani = animation.ArtistAnimation(fig, ims, interval=250, repeat_delay=1000, blit=True)

ani.save("outputs/mnist.gif", writer="imagemagick")
```
